```python
df["Month"]=df["Order Date"].str[3:5]
df
```

```python



aylar = {
    "Ocak": 1, "Şubat": 2, "Mart": 3, "Nisan": 4,
    "Mayıs": 5, "Haziran": 6, "Temmuz": 7, "Ağustos": 8,
    "Eylül": 9, "Ekim": 10, "Kasım": 11, "Aralık": 12
}

df["Month"] = df["Month"].replace(aylar).astype(int)

```

```python
#df["Month"]=[Aylar[i] for i in df["Month"]]   #BU HOCANIN YAPTIGI ŞEKİL ALTTAKİ BENİM Kİ
df["Month"]=df["Month"].replace(aylar)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%d/%m/%Y")  #TO DATEMTME KULLANINCA FORMAT KISMINI UNUTMA!
df
```

```python
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="%d/%m/%Y")
df
```

```python
df["Year"]=df["Order Date"].dt.year   #DT.YEAR YIL TARİHİNİ ALMAMIZI SAGLIYOR!
df
```
